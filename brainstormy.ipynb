```python
def genNumFilter(format, r = range(1,6)):
    return [format.format(num) for num in r]
```

```python
import pandas as pd

data5 = pd.read_html('data/five.html')[0]

ftCols = genNumFilter('ft.of.{}', r=range(2,6))
data5[ftCols] = data5[ftCols].transform(lambda s: s.map(lambda x: int(x.replace(' ','').replace('Ft',''))))

data5
```

```python
data5withWinnings = data5[data5.year > 1997]
data5withWinnings
keys = data5.keys()
winnerData = data5withWinnings[[key for key in keys][3:-5]]
winnerData
```
